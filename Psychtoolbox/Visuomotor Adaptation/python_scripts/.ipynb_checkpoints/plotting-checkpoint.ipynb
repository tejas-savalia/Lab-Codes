```python
import scipy.io
import numpy
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

# Take Note!
mat/data.mat are pilot 3

mat2/data2.mat are pilot 2

mat3/data3.mat are pilot 1

```python
mat = scipy.io.loadmat('data.mat')
mat2 = scipy.io.loadmat('data2.mat')
mat3 = scipy.io.loadmat('data3.mat')
```

```python
mtq = numpy.quantile(mat['mt'], numpy.array([0, 0.1, 0.3, 0.5, 0.7, 0.9, 1]))
```

```python
rmspoints = list()
mtpoints = list()
for i in range(len(mtq)):
    rmspoints.append(mat['rmse'][numpy.where(mat['mt']<mtq[i])[0], numpy.where(mat['mt']<mtq[i])[1]])
    mtpoints.append(mat['mt'][numpy.where(mat['mt']<mtq[i])[0], numpy.where(mat['mt']<mtq[i])[1]])
    
```

```python
def qplots(rmspoints, mtpoints, qno):
    plt.hist2d(rmspoints[qno], mtpoints[qno])
    
```

```python
qplots(rmspoints, mtpoints, 6)
```

```python
def ztrans(data):
    mt = data['mt']
    rmse = data['rmse']
    for i in range(10):
        mt[i] = (mt[i] - numpy.mean(mt[i]))/numpy.std(mt[i])
        rmse[i] = (rmse[i] - numpy.mean(rmse[i]))/numpy.std(rmse[i])
    return (mt, rmse)
```

```python
def block_plot(block, data):

    mt, rmse = ztrans(data)
    plt.hist2d(mt[block], rmse[block], bins = 10)
    return
```

```python
interact(block_plot, block = widgets.IntSlider(min=0,max=9,step=1,value=0), data = widgets.ToggleButtons(options = [mat, mat2, mat3]))
```
